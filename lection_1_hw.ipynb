### Задание 1

Изобразите с помощью matplotlib. Изобразите точку x + 2y + 3z. Найдите угол между векторами x, y и x, z.

```python
x = np.array([1,1])
y = np.array([2, 0])
z = np.array([0,2])
```

```python
#кажется, достаточно это так записать, не прибегая к каким-то специальным методам из numpy

wf = x + 2 * y + 3 * z
wf
```

```python
plt.scatter(wf[0],wf[1])
```

```python
#просто попытался построить стрелочки для трех этих векторов на основе примера из какой-то предыдущей лекции 

ax = plt.axes()

plt.xlim( [0, 5] )
plt.ylim( [0, 4] )

ax.arrow( 0, 0, x[0], x[1], head_width=0.1, head_length=0.2, fc='k', ec='k' )
ax.arrow( 0, 0, y[0], y[1], head_width=0.1, head_length=0.2, fc='k', ec='k' )
ax.arrow( 0, 0, z[0], z[1], head_width=0.1, head_length=0.2, fc='k', ec='k' )

plt.show()
```

```python
# эту функцию я (не очень) честно скопировал из той же лекции, но кажется, хотя бы примерно понял, как она работает

def cosine( a, b ):
    """
    Подсчет косинуса угла между векторами a, b по их координатам
    """
    
    # длины векторов
    aLength = np.linalg.norm( a )
    bLength = np.linalg.norm( b )
    
    return np.dot( a, b ) / ( aLength * bLength )
```

```python
np.arccos( cosine( x, y ) ) * 360 / 2 / np.pi
```

```python
#Угол для x и z попытался получить чуть иначе 
#(не то чтобы не пришлось гуглить, но с помощью презентации и лекции как-то уложил в одно)

xlen = np.sqrt(sum(x*x))
zlen = np.sqrt(sum(z*z))
```

```python
cosinus = np.dot(x,z)/(xlen*zlen)
cosinus
```

```python
np.arccos(cosinus) * 360 / 2 / np.pi
```

```python
m1 = np.array([[2, 2],
               [1, 3]])

m2 = np.array([[4, 1, -1],
               [1, 4, -1],
               [-1, -1, 4]])
```
